```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

dataset = [
    ["Milk", "Cola", "Cookie", "Bread", "Jam"],
    ["Ramen", "Kimchi", "Rice"],
    ["Ramen", "Snack", "Bread", "Jam"],
    ["Rice", "Kimchi"],
    ["Cola", "Bear"],
    ["Cola", "Cookie", "Bread"],
    ["Milk", "Bread", "Jam"],
    ["Ramen", "Cheese", "Ham"],
    ["Bear", "Snack", "Cheese", "Cookie"],
    ["Cookie", "Bread", "Jam", "Snack"],
]

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets["length"] = frequent_itemsets["itemsets"].apply(lambda x: len(x))
frequent_itemsets
```

#### üíÅ association_rules
The `association_rules` function is specifically used for generating association rules from frequent itemsets. These are rules that indicate how certain items in a dataset are associated (or frequently co-occur) with each other. This function is commonly used in market basket analysis, where retailers are interested in understanding the relationships between different products purchased together.


```python
rules1 = association_rules(
    df=frequent_itemsets,
    metric="support",
    min_threshold=0.4,
)
rules1
```

```python
rules1 = association_rules(
    df=frequent_itemsets,
    metric="support",
    min_threshold=0.4,
)
rules1
```
