```python
data = read_excel("04_data1.xlsx")
data
```

```python
data = data.set_index("area")
data
```

```python
data1 = data.loc[
    ["Seoul", "Busan", "Daegu"],
    ["2015_good", "2018_good", "2020_good"],
]
data1
```

```python
data1.index, data1.columns
```

```python
import numpy as np
import matplotlib.pyplot as plt


species = data1.index

x = np.arange(len(species))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(nrows=1, ncols=1, layout="constrained", figsize=(12, 8))

for index, column in enumerate(data1.columns):
    offset = width * index
    rects = ax.bar(x + offset, data1[column], width, label=column)
    ax.bar_label(rects, padding=3)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel("Length (mm)")
ax.set_title("Penguin attributes by species")
ax.set_xticks(x + width, species)
ax.legend(loc="upper left", ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

```python
import numpy as np

index = np.arange(7)

plt.figure(figsize=(20, 4))

data6 = data.loc["Seoul":"Ulsan", "2015_good":"2020_good"]
for year in range(2015, 2021):
    year = str(year) + "_good"
    chartdata = data6[year]
    plt.bar(index, chartdata, width=0.15, label=year)
    index = index + 0.15

plt.xlabel("Area")
plt.ylabel("Micrometer")
plt.xticks(index - 0.53, data6.index)
plt.title("2015&2020 Fine Dust(pm10) Good Line Graph")
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd

data = read_csv("04_data2.csv")
data
```

```python
versicolor_SepalWidth = data.loc[data.Species == "versicolor", ["Species", "SepalWidth"]]
virginica_SepalWidth = data.loc[data.Species == "virginica", ["Species", "SepalWidth"]]

plt.figure(figsize=(10, 6))
plt.hist(versicolor_SepalWidth["SepalWidth"], bins=5, alpha=0.5, label="versicolor", color="r")
plt.hist(virginica_SepalWidth["SepalWidth"], bins=5, alpha=0.5, label="virginica", color="b")

plt.xlabel("Sepal")
plt.ylabel("Width")
plt.title("Iris Petal Group Histogram")
plt.legend()
plt.grid()
plt.show()
```
