```python
df = read_csv("abalone.csv", header=None)
df
```

```python
df.columns = ["Sex", "Length", "Diam", "Height", "Whole", "Shucked", "Viscera", "Shell", "Rings"]
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Sex"] = le.fit_transform(df["Sex"])
```

```python
y = df["Rings"]
X = df.drop("Rings", axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
ly_preds = model.predict(X_test)

rmse = np.sqrt(mean_squared_error(ly_preds, y_test))
print(f"평균제곱근오차: {rmse:.4f}")
```

```python
df = read_csv("bezdekIris.csv")
df
```

```python
df.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["class"] = le.fit_transform(df["class"])
```

```python
df
```

```python
y = df["class"]
X = df.drop("class", axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print("\n--- Logistic Regression Classifier ---")
print(f"accuracy score: {accuracy_score(y_test, pred_lr):.4f}")
print(f"confusion matrix:\n {confusion_matrix(y_test, pred_lr)}")
```
