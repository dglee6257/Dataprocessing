```python
import pandas as pd
```

```python
df = pd.read_csv('D:/datatest/abalone.csv', header = None)
```

```python
df
```

```python
df.columns = ['Sex', 'Length', 'Diam', 'Height', 'Whole', 'Shucked', 'Viscera', 'Shell', 'Rings']
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
```

```python
y = df['Rings']
X = df.drop('Rings',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression()

model.fit(X_train, y_train)

```

```python
from sklearn.metrics import mean_squared_error

```

```python
import numpy as np
```

```python
ly_preds = model.predict(X_test)

mse = np.sqrt(mean_squared_error(ly_preds, y_test))

print('평균제곱근오차', mse)
```

```python
df = pd.read_csv('D:/datatest/bezdekIris.csv')
df
```

```python
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['class'] = le.fit_transform(df['class'])
```

```python
df
```

```python
y = df['class']
X = df.drop('class',axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=42)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```
