```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import seaborn as sns
```

```python
df = pd.read_csv('D:/datatest/test1026.csv')
df
```

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
df['dropoff_datetime'] = pd.to_datetime(df['dropoff_datetime'])
```

```python
df['weekday'] = df.pickup_datetime.dt.day_name()
df['month'] = df.pickup_datetime.dt.month
df['weekday_num'] = df.pickup_datetime.dt.weekday
df['pickup_hour'] = df.pickup_datetime.dt.hour
```

```python
fig = plt.figure(figsize=(12, 12))
plt.title('Passenger per Hour using Count Plot Graph')
sns.countplot(x='passenger_count', hue='vendor_id', data=df)
```

```python
new_df = df.loc[:, ['pickup_hour', 'passenger_count', 'vendor_id', 'month']]
new_df
```

```python
plt.figure(figsize=(13,10))
plt.title('Heat Map Graph')

cor_df = new_df.corr()
upp_mat = np.triu(cor_df)

sns.heatmap(cor_df, annot=True, cmap='RdYlGn', mask=upp_mat)
```
