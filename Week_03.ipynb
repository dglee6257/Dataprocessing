```python
import datetime


def to_datetime(st):
    # Remove the non-numeric characters from the string
    clean_string = "".join(filter(str.isdigit, st))

    # Convert the cleaned string to a datetime object
    date_object = datetime.datetime.strptime(clean_string, "%Y%m")

    # Return Timestamp object
    return pd.to_datetime(date_object)
```

```python
to_datetime("2016년12월")
```

```python
data4["구분"].apply(to_datetime)
```

```python
data4["Date"] = data4["구분"].apply(to_datetime)
data4
```

```python
data4["Year"] = data4.Date.dt.year
data4["Month"] = data4.Date.dt.month
```

```python
data4
```

```python
# 전체 데이터 통계량
data4.describe()
```

```python
d_column = data4.columns[1:4]
d_column
```

```python
# 2014년부터 3년동안의 월별 합계
(
    data4.groupby("Month")[d_column].agg(
        "sum",
        numeric_only=True,
    )
)
```

```python
# 2014년부터 3년동안의 월별 평균
(
    data4.groupby("Month")[d_column].agg(
        "mean",
        numeric_only=True,
    )
)
```

```python
# 2014년부터 3년동안의 연별 합계
(
    data4.groupby("Year")[d_column].agg(
        "sum",
        numeric_only=True,
    )
)
```

```python
# 2014년부터 3년동안의 연별 평균
(
    data4.groupby("Year")[d_column].agg(
        "mean",
        numeric_only=True,
    )
)
```
