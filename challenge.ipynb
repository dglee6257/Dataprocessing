```python
import pandas as pd
drama = pd.read_csv('D:/datatest/top100_kdrama.csv')
drama
```

```python
d = (drama['Synopsis'].str.contains('Korea') | drama['Synopsis'].str.contains('America'))
korea_america =  drama[d]
korea_america
```

```python
korea_america['Aired On'].str.strip().str.split(',')
```

```python
korea_america['Aired On'].str.strip().str.split(',').apply(lambda x: x[-1])
```

```python
korea_america['Day'] = korea_america['Aired On'].str.strip().str.split(',').apply(lambda x: x[-1])
korea_america
```

```python
count = korea_america.groupby('Day').size()
count
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 10))
plt.bar(count.index, count, color='g')
plt.title('number of drama per weekday')
plt.xlabel('Day')
plt.ylabel('count')
plt.grid()
plt.show()
```
