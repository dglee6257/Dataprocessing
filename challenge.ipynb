```python
import pandas as pd
drama = pd.read_csv('D:/datatest/top100_kdrama.csv')
drama
```

```python
d = (drama['Synopsis'].str.contains('Korea') | drama['Synopsis'].str.contains('America'))
korea_america =  drama[d]
korea_america
```

```python
aired_day = korea_america['Aired On'].str.strip().str.split(', ')
aired_day
```

```python
total_day = []
for days in aired_day:
    total_day = total_day + days
total_day    
```

```python
from collections import Counter

count = Counter(total_day)
count
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 10))
plt.bar(count.keys(), count.values(), color='g')
plt.title('Number of dramas per weekday', fontsize=18)
plt.xlabel('Day')
plt.ylabel('count')
plt.grid()
plt.show()
```
